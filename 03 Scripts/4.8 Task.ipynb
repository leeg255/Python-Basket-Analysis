# 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Create column 'max_order' which contains the grouping and transformation
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Create flags
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Test
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean']})
```

# 6. Create a spending flag for each user based on the average price across all their orders using the following criteria:

```python
# create average user price
```

```python
ords_prods_merge['average_user_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.average)
```

```python
ords_prods_merge['average_user_price'].head(15)
```

```python
# create flags
```

```python
ords_prods_merge.loc[ords_prods_merge['average_user_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_user_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# test the flags
```

```python
ords_prods_merge[['user_id', 'average_user_price', 'spending_flag']].head(25)
```

# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# create order frequency
```

```python
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge['order_frequency'].head(15)
```

```python
# create flags
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# test the flags
```

```python
ords_prods_merge[['user_id', 'order_frequency', 'frequency_flag']].head(15)
```
