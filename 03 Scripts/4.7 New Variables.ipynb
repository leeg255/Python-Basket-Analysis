```python
df = merged_ords_prods[:1000000]
```

```python
df.shape
```

```python
# define function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
          return 'Mid range product'
    elif row['prices'] > 15:
          return 'High range product'
    else: return 'Not enough data'
```

```python
# apply function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# find max

df['prices'].max()
```

# if statement with loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

# 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column.

```python
altered_result = []

for value in merged_ords_prods['orders_day_of_week']:
    if value == 0:
        altered_result.append('Busiest days')
    elif value == 1:
        altered_result.append('Busiest days')
    elif value == 3:
        altered_result.append('Slowest days')
    elif value == 4:
        altered_result.append('Slowest days')
    else:
        altered_result.append('Regularly busy')
```

```python
# The top two values are 0 and 1, and the two lowest values are 3 and 4.
```

```python
altered_result
```

# 3. Check the values of this new column for accuracy.

```python
merged_ords_prods['busiest_day'] = altered_result
```

```python
merged_ords_prods['busiest_day'].value_counts(dropna = False)
```
